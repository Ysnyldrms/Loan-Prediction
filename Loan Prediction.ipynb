```python
df["Credit_Score"] = df["Credit_Score"]**2                    # I take squared of Credit Score. Thus we increase the impact of Credit Score(The impact of Credit Score for Price is increased). Why I did: Because most corralated value is Credit Score
df["Annual_Income"] = df["Annual_Income"]**2                  # I take squared of Annual Income. (The reason is same)
df["New_Current_Loan_Amount"] = df["New_Current_Loan_Amount"]**2     # I take squared of Lan Amount. (The reason is same)
```

```python
df["New_Current_Loan_Amount"] = df["New_Current_Loan_Amount"].astype("category")          # I changed the type of new current loan amount as category. 
```

```python
df = df.sample(5000)             # We get 5000 random sample from data frame. Because data is big(training takes a lot of time).
```

```python
x = df[["Annual_Income", "Credit_Score","New_Current_Loan_Amount"]]
```

```python
x = pd.get_dummies(x, drop_first = True )
```

```python
y = df["Loan_Status"]
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
g = GaussianNB()
b = BernoulliNB()
```

```python
g.fit(x,y)
```

```python
b.fit(x,y)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
tahmin1 = g.predict(x)        
tahmin2 = b.predict(x)            
```

```python
accuracy_score(tahmin1,y)
```

```python
accuracy_score(tahmin2,y)
```

```python
precision_score(tahmin2,y)
```

```python
confusion_matrix(tahmin2,y)    
```

```python
confusion_matrix(tahmin1,y)
```

```python
print(classification_report(tahmin1,y))
```

```python
print(classification_report(tahmin2,y))
```
